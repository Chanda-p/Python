Convert age from months to years and save in a new column. Create a new column 'age_converted'

```python
def c_convert(val1,val2):
  val_converted = val1/12
  ratio = val1/val2
  return [val_converted,ratio]
```

Joint Probability

```python
pd.crosstab(index = cars_data2['Automatic'],
            columns = cars_data2['FuelType'],
            normalize = True,
            dropna = True)
```

```python
# Probability of a car having a manual gearbox and CNG Fuel is 0.01
# Probability of a car having a manual gearbox and Diesel Fuel is 0.107
# Probability of a car having a manual gearbox and Petrol Fuel is high at 0.83
# Probability of a car having a automatic gearbox and CNG Fuel is 0
# Probability of a car having a automatic gearbox and Diesel Fuel is 0
# Probability of a car having a automatic gearbox and Petrol Fuel is 0.05

```

Marginal Probability

```python
pd.crosstab(index = cars_data2['Automatic'],
            columns = cars_data2['FuelType'],
            margins = True,
            normalize = True,
            dropna = True)
```

```python
# Probability of a car having a manual gearbox when fuel type are CNG or Diesel or Petrol Fuel is 0.95
# Probability of a car having a fueltype as petrol when gearbox is manual or Automatic is 0.88

```

Conditional Probability

```python
pd.crosstab(index = cars_data2['Automatic'],
            columns = cars_data2['FuelType'],
            margins = True,
            normalize = 'index',
            dropna = True)
```

```python
# when the gearbox is manual type, the probability of getting CNG is 0.01, Diesel is 0.11 and Petrol is 0.87
```

To check whether missing numericals should be imputed with mean or median, we first inspect the dataframe.

```python
cars_data2.describe()
```

Difference between mean and 50% of the 'Age' variable is not much. Thus we can consider mean value.
Difference between mean and median of KM is high, thus we can consider median value.
Mean will be used for filling up HP.

Find the mean of Age

```python
cars_data2['Age'].mean()
```

```python
cars_data2['Age'].fillna(cars_data['Age'].mean(), inplace = True)
```

```python
cars_data2['KM'].fillna(cars_data['KM'].median(), inplace = True)
```
